```python
raw_data = graphlab.SFrame.read_csv('Normal/raw_data.csv')
```

```python
train_data, test_data = raw_data.random_split(0.8, seed=0.3)
```

```python
all_features = train_data.column_names()
```

```python
svm_model = graphlab.svm_classifier.create(train_data, target='flag', features=all_features,
                                          validation_set = test_data, max_iterations = 20)
```

```python
svm_model.get('coefficients').print_rows(num_rows=20)
```

```python
svm_model.evaluate(test_data)
```
